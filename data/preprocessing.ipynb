Annotation 파일 합치기

```python
import pandas as pd
import os
import glob
```

```python
input_path = '/home/ykyoo/yeonk/refined/data/anno/'
output_file = '/home/ykyoo/yeonk/refined/data/anno/anno_total.csv'
```

```python
file_list = glob.glob(os.path.join(input_path, '*.csv'))
file_list
```

```python
with open(output_file, 'w') as write_file:
	for i, file_name in enumerate (file_list):
		with open(file_name, 'r') as original_file: # 파일 읽어오기
			n = 0
			while True:
				line = original_file.readline()
				if n == 0: # 모든 csv 파일 릴레이션 삭제
				# if i != 0 and n == 0: # 첫 번쨰 파일이 아니고, 첫 번쨰 라인인 경우 (릴레이션)
					n += 1
					continue
				if not line:
					print(f'File[{i}] Succeed..!')
					break
				write_file.write(line)
			n += 1

print('File Mergin Succeed..!')
```

text data 붙이기

```python
from tqdm.notebook import tqdm
import pandas as pd
import os
```

```python
# Sess_1.txt 파일 생성은 paste *.txt > Sess.txt 로 (Linux)
a = pd.read_csv('/home/ykyoo/yeonk/refined/data/Sess.txt', sep='\t', encoding='CP949')
a
```

```python
# Column 명 데이터 변경
column_name = ['data']
data = a.values.tolist()
new = pd.DataFrame(data, columns=column_name)
new.loc[-1] = ['한번 드셔보세요. 맛있을 거에요.']
new.index = new.index + 1
new = new.sort_index()
print(new)
```

```python
# text data 따로 저장
new.to_csv('/home/ykyoo/yeonk/refined/data/text_data_20.csv', index=False)
```

```python
# file 이름명 파일 생성
files = os.listdir('/home/ykyoo/yeonk/refined/data/text_data/')
file_list = []
for i in files:
    file_list.append(i)

file_list = pd.DataFrame(file_list)
file_list_1=file_list.sort_values(by=0)
file_list_1.to_csv('/home/ykyoo/yeonk/refined/data/file_list_20.csv', index=False)
```
